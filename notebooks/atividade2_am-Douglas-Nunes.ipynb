## __Enunciado:__
<font size=3>

O conjunto de dados [Linnerud](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_linnerud.html) é um pequeno conjunto clássico usado em Estatística e Aprendizado de Máquina, composto por **20 observações** que relacionam **3 variáveis de exercício físico** (_Chins_ - flexões, _Situps_ - abdominais, _Jumps_ - saltos) com **3 variáveis fisiológicas** (_Weight_ - peso, _Waist_ - cintura, _Pulse_ - frequência cardíaca). Seu principal objetivo é estudar a correlação entre atividade física e medidas corporais.

### __1. Criar um dataframe:__
Importe o conjunto da biblioteca `sklearn`
```python
from sklearn.datasets import load_linnerud
```
<br/>

e crie o **dataframe** usando o `pandas` a fim de obter uma tabela com 6 colunas &mdash; 3 exercícios e 3 variáveis fisiológicas.

P.S: Se basei na aula 3, onde importamos o conjunto de dados de flores de íris.

````python
# Importando a biblioteca pandas
import pandas as pd

# Criando o Dataframe
from sklearn.datasets import load_linnerud

# Importando a biblioteca marplotlib
import matplotlib.pyplot as plt
````

````python
# Criando o datset
linnerud = load_linnerud()

# Criacao do dataframe com 6 colunas
dataframe_exercicios = pd.DataFrame(
    data = linnerud.data,
    columns = linnerud.feature_names
)

dataframe_fisiologia = pd.DataFrame(
    data = linnerud.target,
    columns = linnerud.target_names
)

#print("DataFrame de exercícios \n", dataframe_exercicios)
#print("\n DataFrame de variáveis fisiológicas \n",dataframe_fisiologia)

# Concatenando os Datafreams
dataframe_total = pd.concat([dataframe_exercicios, dataframe_fisiologia], axis = 1)
print(dataframe_total.head())
````

### __2. Análise de dados:__
<font size=3>

Usando os próprios recursos do _dataframe_ calcule:

````python
# A média e desvio padrão dos pesos dos candidatos:
# Média e desvio padrão dos pesos (Weight)

# Calculando a média
media_peso = dataframe_total["Weight"].mean()

# Calculando o desvio padrao
desvio_padrao_peso = dataframe_total["Weight"].std()

print(f"\nMédia do peso:", media_peso)
print(f"Desvio padrão do peso:", desvio_padrao_peso)
````

````python
# Calcule o Intervalo Interquartil do números de saltos ("Jumps") dos candidatos:
# Intervalo interquartil dos saltos (Jumps)

# Calculando o intevalo interquartil dos saltos
terceio_interquartil = dataframe_total["Jumps"].quantile(0.75)
primeiro_interquartil = dataframe_total["Jumps"].quantile(0.25)
intervalo_interquartil = terceio_interquartil - primeiro_interquartil
print("\nIntervalo Interquartil (IQR) dos saltos (Jumps):", intervalo_interquartil)
````

### __3. Plotagem:__
<font size=3>

Utilizando a biblioteca $\texttt{matplotlib}$ faça os seguientes gráficos:

````python
# Extraia os valores de flexões (Chins) e pulso (Pulse) no formato de listas
# para fazer um scatter plot da relação flexão vs pulso. Não esqueça de dar os
# nomes aos eixos x e y.

# Extraindo os valores no formto de lista
chins_x = dataframe_total["Chins"].tolist()
pulse_y = dataframe_total["Pulse"].tolist()

plt.scatter(chins_x, pulse_y, color="red")
plt.xlabel("Chins (Flexões)")
plt.ylabel("Pulse (Frequência Cardíaca)")
plt.title("Relação entre Flexões e Frequência Cardíaca")
plt.show()
````

````python
# Plote um histograma da distribuição do tamanho das cinturas (Waist) dos candidatos:

# Histograma da cintura (Waist)
plt.hist(dataframe_total["Waist"], bins=8, color="blue", edgecolor="black")
plt.xlabel("Cintura (Waist)")
plt.ylabel("Frequência")
plt.title("Distribuição do Tamanho da Cintura")
plt.show()
````
